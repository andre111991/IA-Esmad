## Models

```python
# Import Machine Learning libraries

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
# random sample of all rows
# df in random order
df1=df.sample(frac=1)

train = df1[:120]
test = df1[120:]
X_train = train.drop('target', axis=1)
X_test = test.drop('target', axis=1)
y_train = train['target']
y_test = test['target']
```

```python
# Create the decision tree

dt = DecisionTreeClassifier(criterion="entropy", max_depth=5, random_state=_random_state)
```

```python
# Train the decision tree
dt=dt.fit(X_train, y_train)
```

```python
_features=df.drop(['target'], axis=1)
_feature_names=_features.columns

plt.figure(figsize=(20,15))
plot_tree(dt, 
          filled=True, # Colors help distinguish the classes
          rounded=True, 
          feature_names=_feature_names, # Column names (Sepal, Petal...)
          class_names=dt.classes_, # Flower names (Setosa, Virginica...)
          fontsize=12)
plt.title("Decision Tree\nHow the model decides what flower is it")
plt.show()
```

```python
# Make a prediction with the Decision Tree
# using the test data
prediction = dt.predict(X_test)
```

```python
# Calculate the prediction accuracy 
acc = accuracy_score(y_test, prediction)
print(f"Accuracy: {acc * 100:.2f}%\n")
```

```python
# Check where the machine made mistakes with a Confusion Matrix
fig, ax = plt.subplots(figsize=(8, 6))
plt.title("Confusion Matrix: Where the machine succeeded and where it hesitated")
cm = confusion_matrix(y_test, prediction)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=dt.classes_)
disp.plot(cmap='Blues', ax=ax)
plt.show()
```
